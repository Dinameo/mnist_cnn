```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import load_model
model = load_model('../model/model.h5')
# img = cv2.imread('number1.png')
# img = cv2.imread('number2.jpg')
img = cv2.imread('test.jpg')
img_cnt = img.copy()

# cvt gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.convertScaleAbs(gray, alpha=1, beta=0)


# threshold
binary_img = cv2.threshold(gray, 0, 1, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]

# find contours
contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = [cnt for cnt in contours if cv2.contourArea(cnt) > 150]

def resize_with_padding(img):
    h, w = img.shape[:2]
    scale = min(28 / w, 28 / h)
    new_w = int(w * scale*0.7)
    new_h = int(h * scale*0.7)
    img_resized = cv2.resize(img, (new_w, new_h))
    x_offset = (28 - new_w) // 2
    y_offset = (28 - new_h) // 2
    res = np.zeros((28, 28), dtype=np.uint8)
    res[y_offset:y_offset + new_h, x_offset:x_offset + new_w] = img_resized
    return res

size_sub = int(np.ceil(np.sqrt(len(contours))))
plt.figure(figsize=(2*size_sub, 2*size_sub))

for i, contour in enumerate(contours, start=1):

    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(img_cnt, (x, y), (x+w, y+h), (0, 255, 0), 2)
    img_roi = binary_img[y:y+h, x:x+w]
    plt.subplot(size_sub, size_sub, i)
    
    res = resize_with_padding(img_roi)

    num = np.argmax(model.predict(np.array([res])), axis=-1)

    plt.axis('off')
    plt.imshow(res, cmap='gray')
    plt.title(f'Detect: {num}')
    i += 1
    
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.title("Root")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.title("Gray")
plt.imshow(gray, cmap='gray')
plt.subplot(2, 2, 3)
plt.title("Threshold")
plt.imshow(binary_img, cmap='gray')
plt.subplot(2, 2, 4)
plt.title("Detect")
plt.imshow(cv2.cvtColor(img_cnt, cv2.COLOR_BGR2RGB))
```
